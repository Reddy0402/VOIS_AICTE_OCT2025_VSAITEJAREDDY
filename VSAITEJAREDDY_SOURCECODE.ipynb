## Load data

### Subtask:
Load the Airbnb data from the provided Excel file into a pandas DataFrame.


**Reasoning**:
Import pandas and load the Excel file into a DataFrame, then display the head and info.



```python
import pandas as pd

file_path = "/content/1730285881-Airbnb_Open_Data.xlsx"
df = pd.read_excel(file_path)

# display(df.head())
# display(df.info())
```

### Subtask: Handle Missing Values

Based on the missing value counts, we will handle missing values in several columns.

```python

numerical_cols_to_fill = ['lat', 'long', 'Construction year', 'price', 'service fee', 'minimum nights', 'number of reviews', 'reviews per month', 'review rate number', 'calculated host listings count', 'availability 365']
for col in numerical_cols_to_fill:
    df[col] = df[col].fillna(df[col].median())


categorical_cols_to_fill = ['NAME', 'host_identity_verified', 'host name', 'neighbourhood group', 'neighbourhood', 'country', 'country code', 'instant_bookable', 'cancellation_policy']
for col in categorical_cols_to_fill:
    df[col] = df[col].fillna(df[col].mode()[0])


df['house_rules'] = df['house_rules'].fillna('Not specified')
df['license'] = df['license'].fillna('Not specified')


print("Missing values after handling:")
print(df.isnull().sum())
print("\nData types after handling missing values:")
print(df.dtypes)
```

## Data Cleaning and Preparation

### Subtask:
Check for missing values and data types.

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Check data types
print("\nData types:")
print(df.dtypes)
```

### Subtask: Handle Missing 'last review' Column

The 'last review' column still contains a large number of missing values. Given this, and to simplify the dataset for analysis, we will drop this column.

```python
# Drop the 'last review' column due to the large number of missing values
df = df.drop('last review', axis=1)

print("\nMissing values after dropping 'last review':")
print(df.isnull().sum())
print("\nData types after dropping 'last review':")
print(df.dtypes)
```

## Geospatial Analysis

### Subtask: Visualize Listing Distribution on a Map

Use the latitude and longitude information to visualize the distribution of Airbnb listings on a map.

```python
import folium
from folium.plugins import HeatMap

# Create a base map centered around the average coordinates of the listings
avg_lat = df['lat'].mean()
avg_long = df['long'].mean()
m = folium.Map(location=[avg_lat, avg_long], zoom_start=10)

# Add a heatmap layer to show the density of listings
heat_data = [[row['lat'], row['long']] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(m)

# Display the map
m
```

### Subtask: Analyze Impact of Verified Identity on Reviews

Compare the average review rate number for verified vs. unconfirmed host identities.

### Subtask: Analyze Correlation between Price and Service Fee

Calculate and visualize the correlation between price and service fee.

```python
correlation_price_service_fee = df['price'].corr(df['service fee'])
print(f"\nCorrelation between Price and Service Fee: {correlation_price_service_fee:.2f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='price', y='service fee', alpha=0.6)
plt.title('Relationship between Price and Service Fee')
plt.xlabel('Price')
plt.ylabel('Service Fee')
plt.show()
```

### Subtask: Analyze Neighborhoods with Highest Average Prices

Identify and visualize the neighborhoods with the highest average listing prices.

```python
# Calculate average price by neighbourhood
average_price_by_neighbourhood = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)

# Display the top 10 neighbourhoods with highest average prices
print("Top 10 Neighbourhoods by Average Price:")
print(average_price_by_neighbourhood.head(10))

# Visualize the top 10 neighbourhoods by average price
plt.figure(figsize=(10, 7))
average_price_by_neighbourhood.head(10).plot(kind='barh', color=sns.color_palette('viridis', 10))
plt.title('Top 10 Neighbourhoods by Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.gca().invert_yaxis() # Invert y-axis to show the highest average price at the top
plt.show()
```
